```python
# enable live video from device camera
cap = cv2.VideoCapture(0)
# we need to run a while loop to capture the frames
while(cap.isOpened()):
    # cap.read() capture from camera and returs 2 values
    # response: is true when it returns a frame otherwise it is false
    # frame: is the image frame
    response, frame = cap.read()
    
    # if we wanna convert the image to gray
    #frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # we can get many info from cap object
    # https://docs.opencv.org/3.4/d4/d15/group__videoio__flags__base.html
    # one of the information is height and width of the image
    print((cap.get(cv2.CAP_PROP_FRAME_WIDTH), cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
    
    # show the video in a window
    cv2.imshow('live', frame)
    # break getting frames from camera when i press q key in the keyboard
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# close the camera
cap.release()
# destroy the window
cv2.destroyAllWindows()
```

```python
# above we have seen how to get a video and view it
# here we will learn how to save it
cap = cv2.VideoCapture(0)
# code of our video type
fourcc = cv2.VideoWriter_fourcc(*'XVID')
# to save the video we will call video writer class
# it requires 4 arguments
# 1 - output file name
# 2 - video encode
# 3 - number of frames per second
# 4 - size of the video - width,height
out = cv2.VideoWriter('output.mp4', fourcc, 20.0, (640,480))
while(cap.isOpened()):
    response, frame = cap.read()
    if response:
        cv2.imshow('live', frame)
        #save the frames
        out.write(frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
# release the output file
out.release()
cv2.destroyAllWindows()
```
