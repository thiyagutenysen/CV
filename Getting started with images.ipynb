```python
import cv2
import matplotlib.pyplot as plt
```

```python
# imread reads the image
# imread gets two arguments 1. picture file name, 2. 1 or 0 or -1
# 1 - loads the color image
# 2 - loads the image in grayscale
# 3 - loads image as such with alpha channel too
# imread returns a numpy array with pixel color code from 0-255
# if it is color image size of image = (height, width, 3)
# if it is color image size of image = (height, width, 1)
image = cv2.imread('Ragnar.jpg',-1)
# show the image in the separate window
# imshow code displays the image and suddenly goes away
cv2.imshow('ragnar loth brok',image)
# to make the image window to stay a while we need to specify some time ie., 5000
# if we give time as 0 instead of 5000, the window stays until we click the close button.
cv2.waitKey(5000)
# close the window after specified time
cv2.destroyAllWindows()
```

```python
# how to print the image in jupyter notebook itself
# cv reads the file in blue,green,red format
# but matlab loads as red,blue,green format
# so we need to convert the colors from bgr to rgb using cv2.cvtColor
# plt.imshow method of matplotlib prints the image in console
print(image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# it only returns blue color values
image = cv2.imread('Ragnar.jpg',0)
print(image)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
image = cv2.imread('Ragnar.jpg',1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
